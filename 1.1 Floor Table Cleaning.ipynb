# Floor Table Cleaning

```python
import pandas as pd
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
floor_info = pd.read_csv("bs_harvest_floors.csv")
```

```python
floor_info.head()
```

```python
# abstract infomation from 'data' for each floor

n_row = floor_info.shape[0]

col_name = list(json.loads(floor_info.data.iloc[0]).keys())
stats_col = col_name.copy()
floor_data_stats  = pd.DataFrame(columns = stats_col)

for i in range(n_row):
    f_info = json.loads(floor_info.data.iloc[i])

    if 'stage' in f_info.keys():
        f_info['stage'] = f_info['stage']['name']
    if 'outline' in f_info.keys():
        f_info['outline'] = str(f_info['outline'])
        
    df = pd.DataFrame(f_info, index = [i])
    floor_data_stats = floor_data_stats.append(df)
```

```python
floor_data_stats.rename(columns={'uuid':'uuid_from_data'}, inplace=True)
```

```python
floor_data_stats
```

```python
# Keep features from original table:
#  ---- 'uuid': unique id of floor plan; 
#  ---- 'property_id': unique id of each building (a property may contain multiple floor plans since multiple-level ); 
#  ---- '_fivetran_synced': created date at system
floor_basic = floor_info[['uuid','property_id', 'harvest_id', 'floor_id', 'stargate_floor_id', 'stargate_property_id', '_fivetran_synced']]

# delete 'uuid' in floor_data_stats (?) not clear
# floor_data_stats = floor_data_stats.drop(['uuid'], axis=1)

# combine two tables to get whole information of floor 
if floor_data_stats.shape[0] == floor_basic.shape[0]:
    floor_df = pd.concat([floor_basic, floor_data_stats], axis=1, join_axes=[floor_basic.index])
else:
    floor_df = pd.concat([floor_basic, floor_data_stats], axis=1, join_axes=[floor_basic.index])
    print('warning! You may check your table dimension')
print('shape of floor_data_stats', floor_data_stats.shape)
print('shape of floor_basic', floor_basic.shape) 
print('shape of floor_df(merged)', floor_df.shape)

```
